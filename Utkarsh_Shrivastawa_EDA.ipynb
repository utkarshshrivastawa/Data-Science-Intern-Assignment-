# Data Visualizations

```python
# Customer Distribution by Region
plt.figure(figsize=(10, 6))
sns.countplot(data=customers, x='Region', palette='viridis')
plt.title('Number of Customers by Region')
plt.xlabel('Region')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

```python
# Customer Signup Trends
customers['SignupYear'] = customers['SignupDate'].dt.year
signup_trends = customers['SignupYear'].value_counts().sort_index()

# Plotting the signup trends graph
plt.figure(figsize=(10, 6))
signup_trends.plot(kind='bar', color='skyblue')
plt.title('Customer Signup Trends by Year')
plt.xlabel('Year')
plt.ylabel('Number of Signups')
plt.xticks(rotation=0)
plt.show()
```

```python
# Distribution of Total Transaction Value
plt.figure(figsize=(10, 6))
sns.histplot(transactions['TotalValue'], bins=30, kde=True, color='orange')
plt.title('Distribution of Total Transaction Value')
plt.xlabel('Total Value (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Product Category Sales
category_sales = transactions.merge(products[['ProductID', 'Category']], on='ProductID')
category_sales_summary = category_sales.groupby('Category')['TotalValue'].sum().sort_values()

plt.figure(figsize=(10, 6))
category_sales_summary.plot(kind='bar', color='salmon')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales Value (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Product Categories Distribution
product_category_distribution = products['Category'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=product_category_distribution.values, y=product_category_distribution.index, palette="viridis")
plt.title('Product Categories Distribution', fontsize=14)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.show()
```

```python
# Revenue by Region
region_revenue = transactions.merge(customers, on='CustomerID').groupby('Region')['TotalValue'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=region_revenue, x='Region', y='TotalValue', palette="viridis")
plt.title('Revenue by Region', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Revenue (USD)', fontsize=12)
plt.show()
```
